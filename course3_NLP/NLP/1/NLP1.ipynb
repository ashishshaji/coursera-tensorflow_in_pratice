```
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
sen=['this is a cat',
     'That is a cat','there is a cat in room']
```

```
tokenizer = Tokenizer(num_words = 100,oov_token="ooo")
tokenizer.fit_on_texts(sen)
word_ind=tokenizer.word_index
# converting to list
seq = tokenizer.texts_to_sequences(sen)

print(word_ind)
print(seq)
```

```
padded = pad_sequences(seq, maxlen=5)
print(padded)
```

```
test_data = [
    'there is a white cat',
    'where is my cat'
]

test_seq = tokenizer.texts_to_sequences(test_data)
print("\nTest Sequence = ", test_seq)

padded = pad_sequences(test_seq, maxlen=6)
print("\nPadded Test Sequence: ")
print(padded)
```
